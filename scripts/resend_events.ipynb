```python
import os
import json
from datetime import datetime

has_more = True
start_after = None
scrape_until_ts = 1721685600

events = []

while has_more:
    command = "stripe events list --color off --live --limit 100 "

    if start_after:
        command += f"--starting-after={start_after} "

    # print(result)

    print(command)

    result = os.popen(command).read()

    # print(result)

    data = json.loads(result)

    events += data["data"]

    if not data["has_more"]:
        has_more = False

    # print(events)

    created = [datetime.fromtimestamp(event["created"]).isoformat() for event in events]
    print("Last scraped timestamp", created[-1])

    start_after = events[-1]["id"]

    if events[-1]["created"] < scrape_until_ts:
        break

    
```
